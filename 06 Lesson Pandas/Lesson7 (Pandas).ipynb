```python
# sqlite3 is used to create a connection to a database which we can then use to generate a DataFrame through a SELECT query.
import pymysql
```

```python
# So first we'll make a connection to a SQLite database file:
conn = pymysql.connect(
    host='localhost',
    port=int(3306),
    user="root",
    passwd='1234',
    db="jumiaa_store",
    charset='utf8mb4')
```

```python
# In Our  SQLite database we have a table called user_details

# By passing a SELECT query and our con, we can read from the user_details table:

df = pd.read_sql_query("SELECT * FROM customers",conn)
df.head()
```

If you have data in PostgreSQL,  or some other SQL server, you'll need to obtain the right Python library to make a connection. For example, psycopg2

### Converting back to a CSV, JSON, or SQL

# How to calculate summary statistics?


![agg](img/agg.png)

Aggregation statistics can be calculated on entire columns or rows

groupby provides the power of the split-apply-combine pattern

value_counts is a convenient shortcut to count the number of entries in each category of a variable



```python
# get dataset # https://github.com/mwaskom/seaborn-data
import seaborn as sns 
titanic= sns.load_dataset('titanic')
titanic.head()
```

### What is the average age of the titanic passengers?

```python
titanic["age"].mean()
```

### What is the median age and ticket fare price of the titanic passengers?

```python
 titanic[["age", "fare"]].median()
```

### The aggregating statistic can be calculated for multiple columns at the same time. 

```python

titanic[["age", "fare"]].describe()
```

```python
#Instead of the predefined statistics, 
#specific combinations of aggregating statistics for given columns can be defined using the DataFrame.agg() method:
titanic.agg({'age': ['min', 'max', 'median', 'skew'],
             'fare': ['min', 'max', 'median', 'mean']})
```

### What is the average age for male versus female titanic passengers?

```python
titanic[["sex", "age"]].groupby("sex").mean()
```

### What is the mean ticket fare price for each of the sex and cabin class combinations?

```python
titanic.groupby(["sex", "pclass"])["fare"].mean()
```

### What is the number of passengers in each of the cabin classes?

```python
titanic["pclass"].value_counts()
```

```python
titanic.groupby("pclass")["pclass"].count()
```
